**Functional equations and inequalities stemming from the known quadrature rules**

**Example 1:** The 2-point rule is called the trapezoidal rule. 
$$\left[\frac{1}{2}f(x)  + \frac{1}{2}f(y)\right] =\frac{F(y) - F(x)}{(y-x)}$$
<br>
$$\left[\frac{1}{2}f(x)  + \frac{1}{2}f(y)\right]\leq \frac{1}{y-x}\int_x^y f(t)dt$$

```python
print("\033[1mFunctional equation result\033[0m")
print("")
fe.PSFE(F(y)-F(x) - (y-x)*(Rational(1/2)*f(x) + Rational(1/2)*f(y)))
print("")
print("\033[1mFunctional inequality result\033[0m")
print("")
fi.general_inequality((1/2)*f(x) + (1/2)*f(y))
```

**Example 2:** The 3-point rule is known as Simpson's rule. 
$$F(y) - F(x) = (y-x)\left[\frac{1}{6}f(x) + \frac{2}{3}f\left(\frac{x+y}{2}\right) + \frac{1}{6}f(y)\right]$$
<br>
$$\left[\frac{1}{6}f(x) + \frac{2}{3}f\left(\frac{x+y}{2}\right) + \frac{1}{6}f(y)\right]\leq \frac{1}{y-x}\int_x^y f(t)dt$$

```python
print("\033[1mFunctional equation result\033[0m")
print("")
fe.PSFE(F(y) - F(x)- (y-x)*( Rational(1/6)*f(x) + Rational(2/3)*f((x+y)*Rational(1/2)) +  Rational(1/6)*f(y)))
print("")
print("\033[1mFunctional inequality result\033[0m")
print("")
fi.general_inequality((1/6)*f(x) + (2/3)*f((x+y)*(1/2)) + (1/6)*f(y))
```

**Example 3:** The 4-point closed rule is Simpson's 3/8 rule. 
$$8\left[F(y) -F(x)\right] = 
(y-x)\left[f(x) + 3f\left(\tfrac{x+2y}{3}\right) + 3f\left(\tfrac{2x+y}{3}\right) +f(y)\right] $$
<br>
$$\tfrac{1}{8}\left[f(x) + 3f\left(\tfrac{x+2y}{3}\right) + 3f\left(\tfrac{2x+y}{3}\right) +f(y)\right]\leq \frac{1}{y-x}\int_x^y f(t)dt$$

```python
print("\033[1mFunctional equation result\033[0m")
print("")
fe.PSFE(8*(F(y)-F(x)) - (y-x)*(f(x) + 3*f((x+2*y)*Rational(1/3))  + 3*f((2*x+y)*Rational(1/3)) +f(y)))
print("")
print("\033[1mFunctional inequality result\033[0m")
print("")
fi.general_inequality((1/8)*(f(x) + 3*f((x+2*y)*(1/3))  + 3*f((2*x+y)*(1/3)) +f(y)))
```
